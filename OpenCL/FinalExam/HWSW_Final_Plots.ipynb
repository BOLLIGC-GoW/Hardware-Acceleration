```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#open the file containing the data
file = open('data.txt','r')

#read all of the data from the file
data_lines = file.readlines()

df = pd.read_csv("data.txt", names = ['dimensions', 'filter_size', 'scheme', 'time'])

print (df)

df = df.values

data512 = []
data1024 = []
data2048 = []

#setting relative times compared to CPU
for i in [0,5,10,15,20,25]:
    temp = float(df[i][3])/100
    for j in range(i,i+5):
        df[j][3] = int(float(df[j][3])/temp)
        data512.append(df[j])

for i in [30,35,40,45,50,55]:
    temp = float(df[i][3])/100
    for j in range(i,i+5):
        df[j][3] = int(float(df[j][3])/temp)
        data1024.append(df[j])

for i in [60,65,70,75,80,85]:
    temp = float(df[i][3])/100
    for j in range(i,i+5):
        df[j][3] = int(float(df[j][3])/temp)
        data2048.append(df[j])
        
data512 = sorted(data512,key=lambda x: x[2])
data1024 = sorted(data1024,key=lambda x: x[2])
data2048 = sorted(data2048,key=lambda x: x[2])

file.close()
```

# 512 x 512 Plots

```python
x_axis = [3,5,7,9,11,13]
CPU = []
openCL_base = []
openCL_float4 = []
openCL_float_local8 =  []
openCL_float_local16 = []

for i in range(0,6):
    CPU.append(data512[i][3])
for i in range(6,12):
    openCL_base.append(data512[i][3])
for i in range(12,18):
    openCL_float4.append(data512[i][3])    
for i in range(18,24):
    openCL_float_local8.append(data512[i][3])
for i in range(24,30):
    openCL_float_local16.append(data512[i][3])
    

legend_name = ["CPU","OpenCL_Base","OpenCL_Float4","OpenCL_Float4_Local16","OpenCL_Float4_Local8"]
plt.figure(figsize=(12,8))
plt.plot(x_axis,CPU, label = legend_name[0], linewidth = 2)
plt.plot(x_axis,openCL_base, label = legend_name[1], linewidth = 2)
plt.plot(x_axis,openCL_float4, label = legend_name[2], linewidth = 2)
plt.plot(x_axis,openCL_float_local8, label = legend_name[3], linewidth = 2)
plt.plot(x_axis,openCL_float_local16, label = legend_name[4], linewidth = 2)
plt.legend(loc = 'upper right')
plt.xlabel('Filter Size')
plt.ylabel('% Kernel Time')
plt.title('512 x 512 Relative Performance')

plt.grid(True)
plt.show()
```
